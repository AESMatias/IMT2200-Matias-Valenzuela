## 2. EDA

## 3. PCA

Veamos ahora cómo usar PCA para encontrar información relevante en las diferentes dimensiones de los datos.

Debemos comenzar escalando los datos como hemos comentado en clases.

```python
from sklearn.preprocessing import StandardScaler

x = breast_dataset.loc[:, features].values
x = StandardScaler().fit_transform(x)
```

```python
x.shape
```

```python
np.mean(x),np.std(x)
```

```python
feat_cols = ['feature'+str(i) for i in range(x.shape[1])]
normalised_breast = pd.DataFrame(x,columns=feat_cols)
```

```python
normalised_breast.tail(10)
```

Ahora transformemeos estos datos que están en 30 dimensiones, a una proyección en 2 dimensiones usando PCA.

```python
from sklearn.decomposition import PCA

pca_breast = PCA(n_components=2)
principalComponents_breast = pca_breast.fit_transform(x)
```

```python
principal_breast_Df = pd.DataFrame(data = principalComponents_breast
             , columns = ['principal component 1', 'principal component 2'])
```

```python
principal_breast_Df.tail()
```

```python
print('Explained variability per principal component: {}'.format(pca_breast.explained_variance_ratio_))
```

```python
plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Componente Principal 1', fontsize=20)
plt.ylabel('Componente Principal 2', fontsize=20)
plt.title("Análisis de Componentes Principales", fontsize=20)
targets = ['Benign', 'Malignant']
colors = ['r', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = breast_dataset['label'] == target
    plt.scatter(principal_breast_Df.loc[indicesToKeep, 'principal component 1']
               , principal_breast_Df.loc[indicesToKeep, 'principal component 2'], c = color, s = 50)
plt.legend(targets,prop={'size': 15})
```

Primero que nada, debemos obtener las tres componentes principales, para esto podemos copiar el código de celdas anteriores, y adaptar levemente para añadir una dimensión más en las componentes, así podremos capturar una mayor varianza de los datos:

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

# Copiamos el mismo código de antes, pero adaptado a 3 dimensiones:
pca_breast_3dim = PCA(n_components=3, random_state=100) # Ahora con 3 dimensiones

principal_components_breast_3dim = pca_breast_3dim.fit_transform(x)

# DF con las tres componentes principales de PCA
principal_breast_3d_df = pd.DataFrame(
    principal_components_breast_3dim,
    columns=["principal component 1", "principal component 2", "principal component 3"]
)

# Varianza de cada componente principal
explained_var = pca_breast_3dim.explained_variance_ratio_
print('Explained variability per principal component: {}'.format(explained_var))

print("\n Head de las tres componentes principales de PCA:")
display(principal_breast_3d_df.head(5))
```

Hacemos un split 80-20 de datos de train y test. Entrenamos el clasificador SVM con las tres componentes principales, y lo validamos con los datos de test. La variable 

**principal_components_breast_3dim = pca_breast_3dim.fit_transform(x)**

que se encarga de obtener los 3 componentes principales (3 dimensiones), en la celda anterior, ya está usando la matriz de características normalizada con **StandardScaler**, importada de **sklearn.preprocessing**, por lo que no necesitamos normalizar nuevamente para llevar las características a "la misma escala".

```python
X_train, X_test, y_train, y_test = train_test_split(
    principal_components_breast_3dim,
    breast.target, # Objetivo a predecir
    test_size=0.2,
    stratify=breast.target,
    random_state=100
)


SVC = SVC(kernel="rbf", C=1.5, random_state=100)
SVC.fit(X_train, y_train)

y_pred = SVC.predict(X_test) # Realizamos las predicciones
```

Ahora, revisamos la exactitud de las predicciones con el método SVM, usando las componentes principales (PCA de 3 dimensiones), y obtenemos las métricas principales (las más relevantes):

```python
from sklearn.metrics import (classification_report, accuracy_score,
                             confusion_matrix as sk_confusion_matrix,
                             ConfusionMatrixDisplay)

accuracy_value = accuracy_score(y_test, y_pred)
print(f"Exactitud de SVM con PCA 3D: {accuracy_value}")

# la función classification_report nos da las métricas relevantes del clasificador SVM
report_dict = classification_report(
    y_test,
    y_pred,
    target_names=["Benign", "Malignant"],
    output_dict=True) # Debe ser dict para luego mostrarlo como DF con display

# Armamos un DF con las métricas relevantes, para mostrarlo mejor con display
# Para armar este DF, en la siguiente variable "metrics_df" tuve que usar AI, porque
# no encontraba la forma directa, era necesario transponer el DF primero.
metrics_df = (pd.DataFrame(report_dict).T.loc[["Benign", "Malignant", "accuracy", "macro avg", "weighted avg"],
           ["precision", "recall", "f1-score", "support"]])

display(metrics_df)
```

Aquí, podemos ver que tomando PCA con 3 dimensiones, la exactitud del método SVM en estas 3 dimensiones es de 0.9473684210526315, lo que quiere decir que las tres componentes principales acaparan más del 90% de la varianza de los datos, por lo que la mayor parte de la información original de los datos, pero con menos dimensiones, lo que ocupa menos espacio y es más rápido de computar. En este último sentido, este método es similar a SVD con PCA en Álgebra Lineal.

#### Kernel elegido:

Elegimos el kernel **RBF** porque nos ha dado muy buenos resultados en comparación con los demás más utilizados, el dataset aparentemente contieen relaciones no lineales entre sus características que permiten identificar los tipos de tumores, así que para que el valor de c y, en consecuencia, el margen del hiperplano se maximice, debemos optar por un kernel más complejo, esto es, un kernel no lineal que logre capturar la complejidad de los datos.

Al obtener buenos resultados para los valores de precision y recall para ambas clases de tumores, podemos esperar que mediante el kernel **RBF**, el hiperplano pueda separar bien los tumores benignos de los malignos.

Para una mejor visualización de los resultados de las métricas relevantes del SVM, podemos crear una visualización con la matriz de confusión:

```python
confusion_matrix = sk_confusion_matrix(y_test, y_pred)

disp = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix,
                              display_labels=["Benign", "Malignant"])

disp.plot(cmap="Blues")
plt.title("Matriz de confusión SVM con kernel RBF sobre PCA en 3 dimensiones")
plt.show()
```

Analizando los resultados, podemos ver que, según la matriz de confusión, hay muy pocos falsos negativos y falsos positivos, la mayoría de los tumores se clasifican correctamente. 

Además, como se mencionó en celdas anteriores, al usar PCA con 3 dimensiones, obtenemos mucha varianza de los datos (cerca del 94%), así que obtenemos mucha representatividad de los datos con el beneficio de hacer una reducción de la dimensionalidad a 3 dimensiones (obtenidas de las tres primeras componentes principales), lo que mejora la velocidad del procesamiento y ahorra espacio.
